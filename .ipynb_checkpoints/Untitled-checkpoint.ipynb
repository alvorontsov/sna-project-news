Let's start with linear model

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
X = df.drop(['url', ' shares'], axis=1)
X.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
```

```python
X.head()
```

```python
y = df[' shares']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
print(X_train.shape, y_train.shape)
```

```python
# fit a model
lm = linear_model.Ridge(alpha=0.5)
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

```python
## The line / model
plt.scatter(y_test, predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.show()
```

```python
print 'Score:', model.score(X_test, y_test)
```

Let's try random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regr = RandomForestRegressor(max_depth=25, random_state=0)
```

```python
regr.fit(X_train, y_train)
```

```python
print 'Score:', regr.score(X_test, y_test)
```

Let's take K-nn regressor

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
neigh = KNeighborsRegressor(n_neighbors=100)
neigh.fit(X_train, y_train)
```

```python
neigh.score(X_test, y_test)
```
